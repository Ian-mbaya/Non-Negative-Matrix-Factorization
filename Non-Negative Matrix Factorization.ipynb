### Model Training & Implementation

#### Simple Linear Regression

```python
# Transposing the datasets for compatibility
A_train = A_train.T
B_train = B_train.T
A_test = A_test.T

# Create a function for fitting = a linear regression model
def fit_linear_regression(X, y):
    # Add a bias term (intercept) to X
    X = np.hstack((np.ones((X.shape[0], 1)), X))

    # Calculate the coefficients using the normal equation
    coefficients = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

    return coefficients

```

```python
#Using the Model to Make predictions on the test data
def predict_linear_regression(X, coefficients):
    # Add a bias term (intercept) to X
    X = np.hstack((np.ones((X.shape[0], 1)), X))

    # Make predictions
    predictions = X.dot(coefficients)

    return predictions
```

## 2.0 Model Validation and Perfomance Evaluation


### Splitting the training data for Cross Validation

```python
def train_split(X_rna, X_adt, validation_size=0.2):
    # Set a constant random seed for reproducibility
    np.random.seed(42)  

    num_samples = len(X_rna)
    num_validation_samples = int(validation_size * num_samples)
    random_indices = np.random.permutation(num_samples)

    X_train_rna = X_rna[random_indices[num_validation_samples:]]
    X_valid_rna = X_rna[random_indices[:num_validation_samples]]
    X_train_adt = X_adt[random_indices[num_validation_samples:]]
    X_valid_adt = X_adt[random_indices[:num_validation_samples]]
    

    return X_train_rna, X_valid_rna, X_train_adt, X_valid_adt
```

```python
rna_80, rna_20, adt_80, adt_20 =train_split(A_train,B_train, validation_size=0.2)
```

### Function to Calculate MSE bettween Predicted Values and Split ADT

```python
def mean_squared_error(y_true, y_pred):
    # Ensuring both input matrices have the same shape
    if y_true.shape != y_pred.shape:
        raise ValueError("Input matrices must have the same shape.")

    # Calculate the squared differences element-wise
    squared_errors = (y_true - y_pred) ** 2

    # Calculate the mean of squared errors
    mse = np.mean(squared_errors)

    return mse
```

#### Calculating The Linear Regression MSE

```python
SLR_MSE = mean_squared_error(adt_20, validation_proteins)
print(SLR_MSE)
```

### Perfoming Dimensionality Reduction

### Finding the coeffiecient using SVD

```python
def find_coefficient_SVD(A, y):
    
    A = A.T
    y = y.T
    
    # Calculate the SVD of A
    U, S, VT = np.linalg.svd(A, full_matrices=False)

    # Pseudo-inverse of S
    S_pseudo_inv = np.diag(1 / S)

    # Calculate the coefficient matrix C that maps A to y
    C = y @ VT.T @ S_pseudo_inv @ U.T

    return C.T

```

```python
coefficient_SVD = find_coefficient_SVD(rna_80, adt_80)
```

#### Using the SVD coefficients to Make predictions on Split Data

```python
def predict_SVD(X, coefficient_SVD):
    
    # Make predictions
    prediction_SVD = X.dot(coefficient_SVD)

    return prediction_SVD
```

```python
prediction_SVD = predict_SVD(rna_20, coefficient_SVD)
```

#### Calculating The SVD MSE

```python
SVD_MSE = mean_squared_error(adt_20, prediction_SVD)
print(SVD_MSE)
```

### Finding the Coefficient using Least Sqaures

```python
def least_squares(A, y):
   
    # Calculate the coefficient matrix W using the least squares method
    W, _, _, _ = np.linalg.lstsq(y, A, rcond=None)

    return W.T
```

```python
coefficient_LS = least_squares(rna_80, adt_80)
```

#### Using the Least Square Coefficients to Make predictions on the Split data

```python

def predict_LS(X, coefficient_LS):

    # Make predictions
    prediction_LS = X.dot(coefficient_LS)

    return prediction_LS
```
